```python
# 이미지 하나씩 출력
import cv2
import numpy as np
import sys
import os

img_list_os = os.listdir('./fig/images/')

# print(img_list_os)
# type(img_list_os)
# img_list_os[1]

img_files = []
for i in img_list_os:
    img_path = './fig/images/' + i
    img_files.append(img_path)

# print(img_files)
# img_files[0]
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN,
                    cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    img = cv2.imread(img_files[idx], cv2.IMREAD_COLOR)

    cv2.imshow('scene', img)

    if cv2.waitKey(3000) == 27: # 3초지나면 다음사진
        break
    
    idx += 1

    if idx >= len(img_files):
        idx = 0

cv2.destroyAllWindows()
```

```python
## 영상복사
img1 = cv2.imread('fig/puppy.bmp')

# call by reference, 원본까지 빨간색으로 변함
# img2 = img1

# 데이터 save, 원본은 그대로, 카피한 img2만 빨간색
img2 = img1.copy()

img2[150:250, 300:400] = (0, 0, 255) # 빨간색(RED) 

cv2.imshow('image1', img1)
cv2.imshow('image2', img2)

cv2.waitKey()
cv2.destroyAllWindows()
```
