```python
# mask 만들기 (threshold(역치) 잘 만들기)
src = cv2.imread('fig/cow.png')
dst = cv2.imread('fig/green.png') 

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
ret, src_mask = cv2.threshold(src_gray, 240, 255, cv2.THRESH_BINARY_INV)  # ret: return value

# if src is None or dst is None or src_gray is None:
#     print('image read failed')
#     sys.exit()   

h, w = src.shape[:2] 
dst_crop = dst[400:400+h, 700:700+w] # Call by Reference -> 원본까지 바뀜
# dst = cv2.resize(dst, (w, h), interpolation = cv2.INTER_AREA)
# print("h = ", h, "w = ", w)
cv2.copyTo(src, src_mask, dst_crop)
cv2.imshow('src', src)
cv2.imshow('dst_crop', dst_crop)
# cv2.imshow('src_gray', src_gray)
cv2.imshow('src_mask', src_mask)
cv2.imshow('dst', dst)


cv2.waitKey()
cv2.destroyAllWindows()
```
